```python
import pandas as pd
spotify = pd.read_csv('https://bcdanl.github.io/data/spotify_all.csv')
spotify
```

Cody Johnson has been one of my favorite artists to listen to recently, so his name caught my eye while looking through the spotify data frame.

```python
artist= spotify.query("artist_name=='Cody Johnson'")
artist
```

```python
artist['playlist_name'].nunique()
```

Cody Johnson's music appeared in 12 different playlists.  

```python
tracks=(
    artist
    .drop_duplicates(subset=['track_name'])
    ['track_name'].nunique()
)
tracks
```

```python
artist['album_name'].nunique()
```

27 different songs were featured, and they were from 5 different albums.

```python
artist['duration_min']=artist['duration_ms']/60000
artist['duration_min'].mean()
```

On average, Cody Johnson's songs are 3.53 minutes long.

```python
total_ms= artist['duration_ms'].sum()
total_ms
```
